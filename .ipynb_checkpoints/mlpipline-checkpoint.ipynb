## Complete machine learning pipline 

### Planes 
Missing_values --> one_hote_encoding -->  Scaling -->  feature_selection -->  Model training 

```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder , LabelEncoder , MinMaxScaler 
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline,make_pipeline 
df = pd.read_csv('titanic-Dataset.csv')
```

```python
df.head()
```

```python
df_new = df.drop(columns= ['PassengerId' , 'Name', 'Ticket','Cabin'])
```

```python
df_new
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_new.drop(columns=['Survived']),df_new['Survived'] , random_state=42 ,test_size=0.2)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
X_test.head()
```

```python
X_train.head()
```

#### For Null values 

```python
X_train.info()
```

```python
##  age and embarked columns have null values 
```

```python
trf1 = ColumnTransformer(
      [('impute_age' , SimpleImputer() , [2]),     # ( name , object , applying on column)
      ('impute_embarked' ,SimpleImputer(strategy='most_frequent'),[6])],remainder='passthrough'
)
```

```python
## data type conversion preprocessing 
trf2 = ColumnTransformer([
    ("ohe" , OneHotEncoder(sparse_output=False,handle_unknown='ignore') , [1,6]) ] ,remainder='passthrough' 
)
```

```python
## Scalling 
trf3 = ColumnTransformer([
    ("scale" , MinMaxScaler(),slice(0,10) )
])
```

```python
## feature selection
from sklearn.feature_selection import SelectKBest,chi2
trf4 = SelectKBest(score_func=chi2,k=5)
```

```python
## train the model 
trf5 = DecisionTreeClassifier()
```

# Create a Pipline 

```python
pip = Pipeline([
    ('trf1' , trf1) ,
    ('trf2' , trf2),
    ('trf3' , trf3),
    ('trf4' , trf4),
    ('trf5' , trf5)
])
```

### Pipeline Vs make_pipeline 
Pipeline requres  naming of steps , make_pipelines does not 
(Same applies to ColumnTransformer vs make_column_transformer)

```python
## Alternate syntax 
## easy to apply make_pipeline pip 
```

```python
pip1 = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# if we are using pipeline without model then use fit_transform
pip.fit(X_train,y_train)
```

 ## EXplore the Pipeline 

```python
pip.named_steps
```

```python
print(pip.named_steps['trf1'])
```

```python
# Display pipeline 
from sklearn import set_config
set_config(display = 'diagram')
```

```python
# predict 
y_pred = pip.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(y_test,y_pred)
```

```python
## cross validation for pipeline 
from sklearn.model_selection import cross_val_score 
cross_val_score(pip,X_train,y_train,cv=10,scoring='accuracy').mean()
```

```python
## Grid search cv for best parameters 
from sklearn.model_selection import GridSearchCV
params = {
    'trf5__max_depth' :[ 1,2,3,4,5,None]
}
```

```python
grid = GridSearchCV(pip,params,cv=5,scoring='accuracy') 
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
